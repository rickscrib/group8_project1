```python
oscar_df = pd.DataFrame(pd.read_csv(".\Resources\oscar.csv", index_col='Unnamed: 0'))
# oscar_df = oscar_df.loc[(0:3),:] # limit to first 6 movies
# oscar_df.head()

random_indices = random.sample(list(oscar_df.index), 5)
random_rows = oscar_df.loc[random_indices]

random_df = pd.DataFrame(random_rows)
random_df
```

```python
base_url = "http://www.omdbapi.com/?"
params = { "apikey" : apikey, "type" : "movie" }

for index, row in random_df.iterrows():
    params["t"] = re.sub('\s+', '+', oscar_df.loc[ index , 'Film' ] ) 
    movie_df = requests.get( base_url , params ).json()
    
    pprint(movie_df)
```
